# REGRESSION TASK

# Step 1: Create Features and Target

```python
X = df.drop("SalePrice", axis=1)

y = df["SalePrice"]
```

# Step 2: Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

# Step 3: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

# Step 4: Import Evaluation Metrics

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
```

# Step 5: Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train_scaled, y_train)

pred_lr = lr.predict(X_test_scaled)

print("Linear Regression")
print("R2 =", r2_score(y_test, pred_lr))
print("MAE =", mean_absolute_error(y_test, pred_lr))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_lr)))
```

# Step 6: Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)

ridge.fit(X_train_scaled, y_train)

pred_ridge = ridge.predict(X_test_scaled)

print("Ridge Regression")
print("R2 =", r2_score(y_test, pred_ridge))
print("MAE =", mean_absolute_error(y_test, pred_ridge))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_ridge)))
```

# Step 7: Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1)

lasso.fit(X_train_scaled, y_train)

pred_lasso = lasso.predict(X_test_scaled)

print("Lasso Regression")
print("R2 =", r2_score(y_test, pred_lasso))
print("MAE =", mean_absolute_error(y_test, pred_lasso))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_lasso)))
```

# Step 8: Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)

dt.fit(X_train, y_train)

pred_dt = dt.predict(X_test)

print("Decision Tree")
print("R2 =", r2_score(y_test, pred_dt))
print("MAE =", mean_absolute_error(y_test, pred_dt))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_dt)))
```

# Step 9: Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

print("Random Forest")
print("R2 =", r2_score(y_test, pred_rf))
print("MAE =", mean_absolute_error(y_test, pred_rf))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_rf)))
```

# Step 10: Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(
    random_state=42
)

gb.fit(X_train, y_train)

pred_gb = gb.predict(X_test)

print("Gradient Boosting")
print("R2 =", r2_score(y_test, pred_gb))
print("MAE =", mean_absolute_error(y_test, pred_gb))
print("RMSE =", np.sqrt(mean_squared_error(y_test, pred_gb)))
```

# Step 11: Create Comparison Table

```python
results = pd.DataFrame({
    'Model': [
        'Linear Regression',
        'Ridge',
        'Lasso',
        'Decision Tree',
        'Random Forest',
        'Gradient Boosting'
    ],
    'R2': [
        r2_score(y_test,pred_lr),
        r2_score(y_test,pred_ridge),
        r2_score(y_test,pred_lasso),
        r2_score(y_test,pred_dt),
        r2_score(y_test,pred_rf),
        r2_score(y_test,pred_gb)
    ]
})

results.sort_values(
    by='R2',
    ascending=False,
    inplace=True
)

print(results)
```

### Short Interpretation

> The results show that **Gradient Boosting Regressor** achieved the highest R² score (**0.918**), making it the best-performing
model for predicting house prices. **Random Forest Regressor** also performed very well with an R² score of **0.913**. **Ridge
Regression**, **Linear Regression**, and **Lasso Regression** produced similar results with R² scores around **0.863**.
**Decision Tree Regressor** had the lowest performance with an R² score of **0.850**. Overall, ensemble methods
(Gradient Boosting and Random Forest) outperformed the other models and explained more than 91% of the variation 
in house prices.

# Hyperparameter Tuning: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid_rf.fit(X_train, y_train)

print("Best Parameters:")
print(grid_rf.best_params_)

print("\nBest Cross Validation Accuracy:")
print(grid_rf.best_score_)
```

# Step 1: Create Dataset for Clustering

```python
import numpy as np

cluster_data = df.select_dtypes(include=np.number)

cluster_data.head()
```

# Step 2: Remove Target Variables

```python
cluster_data = cluster_data.drop(
    columns=['SalePrice', 'PriceCategory'],
    errors='ignore'
)
```

# Step 3: Scale the Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(cluster_data)
```

# Step 4: Find Best K Using Elbow Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(scaled_data)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

kmeans_labels = kmeans.fit_predict(scaled_data)

df['KMeans_Cluster'] = kmeans_labels
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(
    scaled_data,
    kmeans_labels
)

print("Silhouette Score:", sil_score)
```

```python
print(df['KMeans_Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8,6))

plt.scatter(
    pca_data[:,0],
    pca_data[:,1],
    c=kmeans_labels
)

plt.title("K-Means Clusters")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=2,
    min_samples=5
)

db_labels = dbscan.fit_predict(scaled_data)

df['DBSCAN_Cluster'] = db_labels
```

```python
print(df['DBSCAN_Cluster'].value_counts())
```

```python
from sklearn.metrics import silhouette_score

if len(set(db_labels)) > 1:
    print(
        silhouette_score(
            scaled_data,
            db_labels
        )
    )
else:
    print("Only one cluster detected")
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_data[:,0],
    pca_data[:,1],
    c=db_labels
)

plt.title("DBSCAN Clusters")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

### Interpretation of the DBSCAN Cluster Plot

> The DBSCAN clustering results show that **most houses belong to one large cluster**, while a few observations appear as isolated points away from the main group. This indicates that the majority of houses in the Ames Housing dataset share similar characteristics, whereas a small number of houses have unusual features and are treated as outliers by the DBSCAN algorithm.

> In the PCA visualization, the dense central region represents the main cluster of properties. The scattered points located far from the center are likely houses with exceptionally high or low values in features such as living area, overall quality, garage size, or sale price. DBSCAN successfully identified these unusual observations without requiring a predefined number of clusters.

> Compared to K-Means, DBSCAN did not separate the data into several clearly distinct groups. Instead, it focused on detecting dense regions and identifying outliers. Therefore, DBSCAN is more useful for anomaly detection in this dataset, while K-Means is generally better for segmenting houses into meaningful groups.

### Short Interpretation (for report table)

> DBSCAN identified one dominant cluster containing most houses and a small number of outliers. This suggests that the dataset is relatively homogeneous, with only a few properties exhibiting unusual characteristics. DBSCAN was effective for outlier detection but less effective than K-Means for creating distinct housing segments.


# Lab 04 – Questions and Answers

## 1. How did you handle missing values and why did you choose those strategies?

Missing numerical values were replaced with the median of each column, while missing categorical values were replaced with the mode. The median is less sensitive to extreme values than the mean, and the mode preserves the most common category. This approach allowed all records to be retained for analysis.

## 2. Explain the difference between regression and classification. How did you transform SalePrice into a classification target?

Regression predicts a continuous numerical value, whereas classification predicts categories or classes. SalePrice was converted into a classification target by comparing each house price to the median sale price. Houses above the median were labeled as HighPrice (1), and those below the median were labeled as LowPrice (0).

## 3. Explain how each model is working.

* Linear Regression: Fits a straight-line relationship between predictors and target.
* Ridge Regression: Similar to linear regression but includes regularization to reduce overfitting.
* Lasso Regression: Uses regularization and can eliminate less important features.
* Decision Tree: Splits data into branches based on feature values.
* Random Forest: Combines many decision trees and averages their predictions.
* Gradient Boosting: Builds trees sequentially, correcting previous errors.
* Logistic Regression: Predicts class probabilities.
* KNN: Classifies based on nearest neighbors.
* SVM: Finds the optimal boundary separating classes.

## 4. Which regression model performed best and why?

Gradient Boosting Regressor performed best with an R² score of approximately 0.918. It achieved the highest predictive accuracy because it learns from previous prediction errors and captures complex relationships in the data.

## 5. How does a decision tree differ from a random forest?

A decision tree uses a single tree structure, while a random forest combines many trees and aggregates their predictions. Random forests generally outperform single trees because they reduce overfitting and improve generalization.

## 6. Why is scaling important for KNN and SVM?

KNN and SVM rely on distance calculations. Features with larger scales can dominate the calculations, leading to biased results. Scaling ensures all variables contribute equally.

## 7. What is the silhouette score?

The silhouette score measures how well data points fit within their assigned cluster compared to other clusters. Values closer to 1 indicate better cluster separation and quality.
